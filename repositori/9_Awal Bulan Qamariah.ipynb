```python
import pandas as pd
import math as m
from math import pi
```

```python
lt_tempat  = {"Derajat" : -7, "menit" : 23, "detik" : 56}
bj_tempat  = {"Derajat" : 110, "menit" : 27, "detik" : 53}
ketinggian = 79
wd = "WIB"
```

```python
def deg_to_des(deg, menit, detik):
    sign            = -1 if deg < 0  else 1
    desimal         = abs(deg)+ (menit / 60) + (detik/3600)

    return desimal * sign 

def format_dms(desimal):
    sign            = -1 if desimal < 0  else 1
    decimal_degrees = abs(desimal)
    degrees         = int (decimal_degrees) * sign 
    minutes         = int ((decimal_degrees - abs(degrees))*60)
    seconds         = (decimal_degrees - abs(degrees) - minutes / 60)* 3600

    return f"{degrees} : {minutes} : {seconds:.2f}\""
```

```python
decimal_lt = deg_to_des(lt_tempat["Derajat"], lt_tempat["menit"], lt_tempat["detik"])
decimal_bj = deg_to_des(bj_tempat["Derajat"], bj_tempat["menit"], bj_tempat["detik"])
```
